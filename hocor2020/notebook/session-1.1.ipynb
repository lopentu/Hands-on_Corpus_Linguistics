# CoreNLP與自動詞意標記 - 前言

## 結構化資料

```
import seaborn as sns
```

```
iris = sns.load_dataset("iris")
sns.set(rc={'figure.facecolor':'white'})
sns.pairplot(iris, hue='species')
```

## 分類器

```
from sklearn.svm import SVC
svm = SVC()
```

```
iris.shape
```

```
data = iris.sample(frac=1)
train = data.iloc[:130,:]
Xcols = "sepal_length,sepal_width,petal_length,petal_width".split(",")
svm.fit(train.loc[:,Xcols], train.species)
```

```
test = data.iloc[130:,:]
svm.score(test.loc[:,Xcols], test.species)
```
